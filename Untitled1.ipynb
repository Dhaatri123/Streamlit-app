```python
# Load your data from Excel

file_path = '/content/The English Tree.xlsx'  # Path to your uploaded file
excel_data = pd.read_excel(file_path, sheet_name=None)  # Load all sheets

# Prepare data
inputs, outputs = [], []
for sheet_name, sheet_data in excel_data.items():
    sheet_inputs = sheet_data['Question'].astype(str).tolist()
    sheet_outputs = sheet_data['Answer'].astype(str).tolist()
    inputs.extend(sheet_inputs)
    outputs.extend(sheet_outputs)

# Combine questions and answers for model input
data = {"text": [f"Question: {q} Answer: {a}" for q, a in zip(inputs, outputs)]}

# Convert to Hugging Face Dataset
train_data = Dataset.from_dict(data)

# Load the tokenizer for gpt2 and set the padding token
tokenizer = AutoTokenizer.from_pretrained("gpt2")
tokenizer.pad_token = tokenizer.eos_token  # Use EOS as padding token

# Tokenization function
def tokenize_function(examples):
    return tokenizer(examples["text"], padding="max_length", truncation=True, max_length=128)

# Tokenize the data
tokenized_data = train_data.map(tokenize_function, batched=True)
```
